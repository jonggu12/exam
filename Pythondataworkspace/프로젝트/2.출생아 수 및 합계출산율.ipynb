```python
import pandas as pd
import matplotlib
matplotlib.rcParams['font.family'] = 'AppleGothic' 
import matplotlib.pyplot as plt
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
plt.plot(df.index,df['출생아 수'])
plt.plot(df.index, df['합계 출산율'])
```

```python
fig, ax1 = plt.subplots(figsize = (10,7))
ax1.plot(df.index,df['출생아 수'], color='#ff812d')

ax2 = ax1.twinx() # x축을 공유하는 쌍둥이 axis

ax2.plot(df.index,df['합계 출산율'],color='#ffd100')
```

```python
fig, ax1 = plt.subplots(figsize = (13,5))
ax1.set_ylabel('출생아 수 (천 명)')
ax1.set_ylim(250,700)
ax1.set_yticks([300,400,500,600])
ax1.bar(df.index,df['출생아 수'], color='#ff812d')

ax2 = ax1.twinx() # x축을 공유하는 쌍둥이 axis
ax2.set_ylabel('합계 출산율')
ax2.set_ylim(0,1.5)
ax2.set_yticks([0,1])
ax2.plot(df.index,df['합계 출산율'],color='#ffd100')
```

```python
fig, ax1 = plt.subplots(figsize = (13,5))
ax1.set_ylabel('출생아 수 (천 명)')
ax1.set_ylim(250,700)
ax1.set_yticks([300,400,500,600])
ax1.bar(df.index,df['출생아 수'], color='#ff812d')
for idx, value in enumerate(df['출생아 수']):
    ax1.text(idx,value + 12,value, ha = 'center')





ax2 = ax1.twinx() # x축을 공유하는 쌍둥이 axis
ax2.set_ylabel('합계 출산율')
ax2.set_ylim(0,1.5)
ax2.set_yticks([0,1])
ax2.plot(df.index,df['합계 출산율'],color='#ffd100')
```

```python
fig, ax1 = plt.subplots(figsize = (13,5))
ax1.set_ylabel('출생아 수 (천 명)')
ax1.set_ylim(250,700)
ax1.set_yticks([300,400,500,600])
ax1.bar(df.index,df['출생아 수'], color='#ff812d')
for idx, value in enumerate(df['출생아 수']):
    ax1.text(idx,value + 12,value, ha = 'center')





ax2 = ax1.twinx() # x축을 공유하는 쌍둥이 axis
ax2.set_ylabel('합계 출산율')
ax2.set_ylim(0,1.5)
ax2.set_yticks([0,1])
ax2.plot(df.index,df['합계 출산율'],color='#ffd100',marker='o',ms=10,
         lw=3,mec='w',mew=2)
for idx, value in enumerate(df['합계 출산율']):
    ax2.text(idx,value + 0.08,value, ha = 'center') # x,y 축 정하고 값넣고 중앙 위치
```
