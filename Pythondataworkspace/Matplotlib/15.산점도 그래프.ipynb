```python
import pandas as pd

df = pd.read_excel('/Users/choejong-gyu/Documents/Pythondataworkspace/Pandas/score.xlsx')
df
```

```python
df['학년'] = [3,3,2,1,1,3,2,2]
df
```

```python
plt.scatter(df['영어'],df['수학'])
plt.xlabel('영어점수')
plt.ylabel('수학점수')
```

```python
import numpy as np

sizes = np.random.rand(8) * 1000
sizes
```

```python
plt.scatter(df['영어'],df['수학'],s=sizes)
plt.xlabel('영어점수')
plt.ylabel('수학점수')
```

```python
sizes = df['학년'] * 500 # 1학년은 500 2학년은 1000, 3학년은 1500

plt.scatter(df['영어'],df['수학'],s=sizes)
plt.xlabel('영어점수')
plt.ylabel('수학점수')
```

```python
sizes = df['학년'] * 500 # 1학년은 500 2학년은 1000, 3학년은 1500

plt.scatter(df['영어'],df['수학'],s=sizes, c=df['학년'],cmap='viridis',alpha=0.3)
plt.xlabel('영어점수')
plt.ylabel('수학점수')
```

```python
plt.figure(figsize=(7,7))
plt.scatter(df['영어'],df['수학'],s=sizes, c=df['학년'],cmap='viridis',alpha=0.3)
plt.xlabel('영어점수')
plt.ylabel('수학점수')
plt.colorbar(ticks=[1,2,3], label ='학년',shrink=0.5, orientation = 'horizontal')
```
