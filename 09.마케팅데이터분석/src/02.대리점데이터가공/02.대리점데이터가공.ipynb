### 5 : 금액 결측치 수정

* 결측치 수정 전

```python
uriage_data.isnull().any(axis=0) # .any(axis=0)각 열에 하나라도 null이 있으면 True
```

```python
uriage_data.info()
```

```python
flg_is_null = uriage_data['item_price'].isnull()
flg_is_null
```

```python
uriage_data.loc[flg_is_null,'item_name'].unique()
```

```python
for trg in uriage_data.loc[flg_is_null, 'item_name'].unique():
    price = uriage_data[(~flg_is_null) & (uriage_data['item_name'] == trg)]['item_price'].max()
    uriage_data['item_price'].loc[(flg_is_null) & (uriage_data['item_name']==trg)] = price
    

uriage_data.head(20)
```

* 결측치 처리 후 값 확인

```python
uriage_data.isnull().any(axis=0)
```

```python
for trg in uriage_data['item_name'].sort_values().unique():
    print(trg + '의 최고가: ' + str(uriage_data.loc[uriage_data['item_name']==trg]['item_price'].max()) +
          ', 최저가: ' + str(uriage_data.loc[uriage_data['item_name']==trg]['item_price'].min(skipna=False)))
# skipna=False : nan값 적용
```

### 7 : 날짜오류를 수정하자

```python
flg_is_serial = kokyaku_data['등록일'].astype('str').str.isdigit()
flg_is_serial
```

```python
flg_is_serial.sum()
```

```python
kokyaku_data.loc[flg_is_serial,'등록일'].astype('float')
```

```python
pd.to_timedelta(kokyaku_data.loc[flg_is_serial, '등록일'].astype('float'), unit = 'D')
```

```python
fromSerial = pd.to_timedelta(kokyaku_data.loc[flg_is_serial, '등록일'].astype('float'), unit = 'D') + pd.to_datetime('1900/01/01')
fromSerial          # 숫자로 된 애들 
```

```python
fromString = pd.to_datetime(kokyaku_data.loc[~flg_is_serial, '등록일']) #날짜로 된 애들
fromString
```

```python
kokyaku_data['등록일'] = pd.concat([fromSerial, fromString])
kokyaku_data
```

```python
kokyaku_data['등록연월'] = kokyaku_data['등록일'].dt.strftime('%Y%m')
rslt = kokyaku_data.groupby('등록연월')['고객이름'].count()
print(rslt)
print(len(kokyaku_data))
```
