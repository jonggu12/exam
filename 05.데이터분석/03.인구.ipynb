```python
import matplotlib.pyplot as plt
from matplotlib import colors as mcolors
colors = dict(mcolors.BASE_COLORS, **mcolors.CSS4_COLORS)
# Sort colors by hue, saturation, value and name.
by_hsv = sorted((tuple(mcolors.rgb_to_hsv(mcolors.to_rgba(color)[:3])), name)
                for name, color in colors.items())
sorted_names = [name for hsv, name in by_hsv]
n = len(sorted_names)
ncols = 4
nrows = n // ncols
fig, ax = plt.subplots(figsize=(9, 8), dpi = 300)
# Get height and width
X, Y = fig.get_dpi() * fig.get_size_inches()
h = Y / (nrows + 1)
w = X / ncols
for i, name in enumerate(sorted_names):
    row = i % nrows
    col = i // nrows
    y = Y - (row * h) - h
    xi_line = w * (col + 0.05)
    xf_line = w * (col + 0.25)
    xi_text = w * (col + 0.3)
    ax.text(xi_text, y, name, fontsize=(10),
            horizontalalignment='left',
            verticalalignment='center')
    ax.hlines(y + h * 0.1, xi_line, xf_line,
              color=colors[name], linewidth=(6))
ax.set_xlim(0, X)
ax.set_ylim(0, Y)
ax.set_axis_off()
fig.subplots_adjust(left=0, right=1,
                    top=1, bottom=0,
                    hspace=0, wspace=0)
plt.show()
```

```python
# 지역의 남녀 성별 비율을 파이

import csv
import matplotlib.pyplot as plt

f = open(r'/Users/choejong-gyu/Documents/05.데이터분석/data/gender (3).csv',encoding='CP949')


data = csv.reader(f)

A = input("지역을 입력하세요: ")

size = []

for row in data:
    if A in row[0]:
        m = 0
        f = 0
        for i in range(101):
            m += int(row[i+3].replace(',', ''))
            f += int(row[i+106].replace(',', ''))
             
        break

size.append(m)
size.append(f)

plt.rc('font',family = 'AppleGothic')
color = ['b','r']

plt.axis('equal')
plt.title(f"{A}의 남녀 성별 인구 분포")
plt.pie(size,labels = ['Male','female'],autopct= '%1.1f%%',colors = color, startangle= 90)



plt.show()
```

```python
# 지역의 남녀 성별 비율을 꺾은 선 그래프로 표현

import csv
import matplotlib.pyplot as plt

f = open(r'/Users/choejong-gyu/Documents/05.데이터분석/data/gender (3).csv',encoding='CP949')


data = csv.reader(f)

A = input("지역을 입력하세요: ")

m = []
f = []

for row in data:
    if A in row[0]:
        for i in range(3,104):
            m.append(int(row[i].replace(',', '')))
            f.append(int(row[i + 103].replace(',', ''))) # 여자는 106부터
             
        break



plt.rc('font',family = 'AppleGothic')
color = ['b','r']


plt.title(f"{A}의 남녀 성별 인구 분포")
plt.plot(m,label = 'Male')
plt.plot(f,label = 'female')
plt.legend()

plt.show()
```

```python
# 연령 별로 남성 인구수에서 여성 인구수를 뺀 차이를 막대 그래프로 표현


import csv
import matplotlib.pyplot as plt

f = open(r'/Users/choejong-gyu/Documents/05.데이터분석/data/gender (3).csv',encoding='CP949')
data = csv.reader(f)




A = input("지역을 입력하세요: ")

result = []

for row in data:
    if A in row[0]:
        for i in range(3,104):
            result.append(int(row[i].replace(',', ''))- int(row[i+103].replace(',', ''))) #양수면 남자 음수면 여자
             
        break





plt.rc('font',family = 'AppleGothic')
plt.title(f"{A}의 남녀 성별 인구 차이 분포")
plt.bar(range(101),result)
plt.show()
```
