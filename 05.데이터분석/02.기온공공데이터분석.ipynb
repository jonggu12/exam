```python
import csv


max_temp = -999 #max_temp와 row[-1]를 비교하면서 더 큰값을 max_temp에 넣음 max_temp < row[-1]   -> max_temp = row[-1]
max_date =''
f = open(r'/Users/choejong-gyu/Documents/05.데이터분석/data/서울.csv','r',encoding='CP949')
data = csv.reader(f,delimiter=',')
header = next(data)


for row in data:
    if row[-1] =='':
        row[-1] = -999 # -999 를 넣어 빈 문자열이 있던 자리라고 표시
    row[-1] = float(row[-1])
    if max_temp < row[-1]:
        max_temp = row[-1]
        max_date = row[0].strip()



f.close()



print(f"최고기온: {max_temp}, 기간: {max_date}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(14)
y = np.sin(x / 2)

plt.step(x, y + 2, label='pre (default)')
plt.plot(x, y + 2, 'o--', color='grey', alpha=0.3)

plt.step(x, y + 1, where='mid', label='mid')
plt.plot(x, y + 1, 'o--', color='grey', alpha=0.3)

plt.step(x, y, where='post', label='post')
plt.plot(x, y, 'o--', color='grey', alpha=0.3)

plt.grid(axis='x', color='0.95')
plt.legend(title='Parameter where:')
plt.title('plt.step(where=...)')
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

file_path = r'/Users/choejong-gyu/Documents/05.데이터분석/data/서울.csv'

with open(file_path, 'r', encoding='CP949') as f:
    data = csv.reader(f, delimiter=',')
    header = next(data)
    result = []

    for row in data:
        if row[-1] != '':
            Month = row[0].split('-')[1]
            if Month == "10":
                row[-1] = float(row[-1])
                Day = int(row[0].split('-')[2])
                if Day == 22:
                    result.append(row[-1])

# Close the file after reading
f.close()

# Plotting the result
plt.figure(figsize=(10, 2), dpi=300)
plt.plot(result, 'r')


plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

file_path = r'/Users/choejong-gyu/Documents/05.데이터분석/data/서울.csv'

with open(file_path, 'r', encoding='CP949') as f:
    data = csv.reader(f, delimiter=',')
    header = next(data)

    high = []
    low = []

    for row in data:
        if row[-1] != '':
            date = row[0].split('-')
            if date[1]=="10" and date[2] == "22":
                    high.append(row[-1])
                    low.append(row[-2])
                
                
         


plt.figure(figsize = (10,2), dpi = 300) # 1인치 = 2.54 센티미터
plt.plot(high, 'blue')
plt.plot(low, 'red')
plt.show()
f.close()
```

```python
# 1월 ~ 12월의 데이터 시각화


import csv
import matplotlib.pyplot as plt

f = open(r'/Users/choejong-gyu/Documents/05.데이터분석/data/서울.csv','r',encoding='CP949')
data = csv.reader(f,delimiter=',')
header = next(data)


month = [[],[],[],[],[],[],[],[],[],[],[],[]]


for row in data:
    if row[-1] != '':
        for i in range(1,13): #1월부터 12월
            if int(row[0].split('-')[1]) == i:
                month[i - 1].append(float(row[-1]))
                
plt.figure(figsize = (10,2), dpi = 300)                       
plt.boxplot(month)
plt.show()  
```
