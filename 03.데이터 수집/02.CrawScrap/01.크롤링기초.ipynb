```python
import requests
from bs4 import BeautifulSoup

res = requests.get("https://search.daum.net/search?w=tot&q=2022%EB%85%84%EC%98%81%ED%99%94%EC%88%9C%EC%9C%84&DA=MOR&rtmaxcoll=MOR")
res.raise_for_status()  # 비정상일 때 종료

html_text = res.text

bs_obj = BeautifulSoup(html_text, "html.parser")

# 영화 순위를 가진 부분을 찾습니다.
rankings = bs_obj.find_all("div", {"class": "ol"})

# 찾은 모든 요소를 순회하며 영화 순위를 출력합니다.
for ranking in rankings:
    movie_list = ranking.find("ul", {"class": "list_movie"})
    if movie_list:
        movies = movie_list.find_all("li")
        for index, movie in enumerate(movies, start=1):
            title = movie.find("a", {"class": "f_link_b"})
            if title:
                print(f"{index}. {title.text}")
```

```python
from bs4 import BeautifulSoup

html_doc = """
<html>
<head>
    <title>Sample Page</title>
</head>
<body>
    <h1>Welcome to Beautiful Soup</h1>
    <p class="content">Beautiful Soup is a Python library for parsing HTML and XML documents.</p>
    <p class="content">It creates a parse tree from page source code that can be used to extract data easily.</p>
</body>
</html>
"""

# HTML 파싱
soup = BeautifulSoup(html_doc, 'html.parser')

# 특정 태그 내용 가져오기
title = soup.title
h1 = soup.h1
paragraphs = soup.find_all('p')

# 내용 출력
print("Title:", title.text if title else "No title found")
print("H1:", h1.text if h1 else "No h1 found")
for p in paragraphs:
    print("Paragraph:", p.text)
```

```python
import requests
from bs4 import BeautifulSoup

url = "https://search.daum.net/search?w=tot&q=2022%EB%85%84%EC%98%81%ED%99%94%EC%88%9C%EC%9C%84&DA=MOR&rtmaxcoll=MOR"

response = requests.get(url)
response.raise_for_status()  # 오류 발생 시 예외 처리

html_content = response.text
soup = BeautifulSoup(html_content, "html.parser")

rankings = soup.find_all(class_ = "type_plural list_exact movie_list") #"type_plural list_exact movie_list" 클래스 모두 rankings에 담기



for ranking in rankings: # 첫번째 for문은 "type_plural list_exact movie_list"가 모두 있는 rankings에서 제목, 누적, 평점의 클래스만 뽑기 위해 사용
    movie_titles = ranking.find_all(class_= "info_tit")
    accumulate = ranking.select(".dl_comm:nth-of-type(4)")
    grade = ranking.find_all(class_ = "dl_comm star_comp_s")
    year_acc = ranking.select(".dl_comm:nth-of-type(3)")
    release = ranking.select(".dl_comm:nth-of-type(2)")

    
    

    for index, (title,acc,gr,year,rel) in enumerate(zip(movie_titles,accumulate,grade,year_acc,release,), start=1):    # 두번째 for문은 제목, 누적, 평점의 클래스 안에서 순위를 정하기 위한 for문

        print(f"{index}){title.text}\n{gr.text}\n{year.text}\n{acc.text}\n{rel.text}")
      
    

```

```python
#이미지 저장 1년

import requests
from bs4 import BeautifulSoup
import urllib

url = "https://search.daum.net/search?w=tot&q=2022%EB%85%84%EC%98%81%ED%99%94%EC%88%9C%EC%9C%84&DA=MOR&rtmaxcoll=MOR"

response = requests.get(url) # url 요청
response.raise_for_status()  # 요청이 실패한 경우 에러 처리

html_content = response.text # response의 텍스트를 모두 불러옴
soup = BeautifulSoup(html_content, "html.parser")

# 이미지 태그 모두 찾기
image_tags = soup.find_all(class_="thumb_img")

# 파일이 저장될 경로 지정
directory = '/Users/choejong-gyu/Desktop/image/'  # 이미지가 저장될 디렉토리명

file_no = 0
for img_tag in image_tags:
    img_url = img_tag['src'] # src의 속성의 값을 가져옴.
    try:
        filename = f"{directory}/image{file_no}.jpg"  # 파일명 지정
        urllib.request.urlretrieve(img_url, filename) # 이미지 저장하는 함수
        file_no += 1
    except Exception as e:
        print(f"Error occurred: {e}") #에러났을 때
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
import requests
import pandas as pd

service = Service()
options = webdriver.ChromeOptions()
driver = webdriver.Chrome(service=service, options=options)
driver.get("https://www.coupang.com/np/categories/194276")
list_elements = driver.find_elements(By.CLASS_NAME, 'baby-product-link')

file_no = 0
directory = '/Users/choejong-gyu/Desktop/image/'  # 원하는 디렉토리로 변경하세요


data_list = []

# 텍스트 및 이미지 크롤링
for index, element in enumerate(list_elements, start=1):
    product_name = element.find_element(By.CLASS_NAME, "name")
    price = element.find_element(By.CLASS_NAME, "price-value")
    rate = element.find_element(By.CLASS_NAME, "rating")
    count = element.find_element(By.CLASS_NAME, "rating-total-count")
    
    # 이미지 저장
    try:
        img_tag = element.find_element(By.TAG_NAME, "img")
        img_url = img_tag.get_attribute("src")
        print(f"{index}) {product_name.text}\n{price.text}원\n{rate.text}점\n평점{count.text}개")
        print(f"이미지 URL: {img_url}")

        if img_url and img_url.startswith("http"):  # URL이 비어 있지 않고 "http"로 시작하는지 확인
            filename = f"{directory}/{file_no+1}등,{product_name.text}.jpg"
            with open(filename, "wb") as img_file:
                img_file.write(requests.get(img_url).content)
            file_no += 1
        else:
            print("잘못된 이미지 URL 또는 빈 URL")
        data_list.append({
            '이름': product_name.text,
            '가격': price.text,
            '평점': rate.text,
            'URL': img_url,
            '파일이름': filename
        })
    except Exception as e:
        print(f"오류 발생: {e}")
        

final = pd.DataFrame(data_list)
final.to_csv('쿠팡식품.csv', encoding='utf-8-sig')
final      
```
