```python
import requests
from bs4 import BeautifulSoup

res = requests.get("https://search.daum.net/search?w=tot&q=2022%EB%85%84%EC%98%81%ED%99%94%EC%88%9C%EC%9C%84&DA=MOR&rtmaxcoll=MOR")
res.raise_for_status()  # ÎπÑÏ†ïÏÉÅÏùº Îïå Ï¢ÖÎ£å

html_text = res.text

bs_obj = BeautifulSoup(html_text, "html.parser")

# ÏòÅÌôî ÏàúÏúÑÎ•º Í∞ÄÏßÑ Î∂ÄÎ∂ÑÏùÑ Ï∞æÏäµÎãàÎã§.
rankings = bs_obj.find_all("div", {"class": "ol"})

# Ï∞æÏùÄ Î™®Îì† ÏöîÏÜåÎ•º ÏàúÌöåÌïòÎ©∞ ÏòÅÌôî ÏàúÏúÑÎ•º Ï∂úÎ†•Ìï©ÎãàÎã§.
for ranking in rankings:
    movie_list = ranking.find("ul", {"class": "list_movie"})
    if movie_list:
        movies = movie_list.find_all("li")
        for index, movie in enumerate(movies, start=1):
            title = movie.find("a", {"class": "f_link_b"})
            if title:
                print(f"{index}. {title.text}")
```

```python
from bs4 import BeautifulSoup

html_doc = """
<html>
<head>
    <title>Sample Page</title>
</head>
<body>
    <h1>Welcome to Beautiful Soup</h1>
    <p class="content">Beautiful Soup is a Python library for parsing HTML and XML documents.</p>
    <p class="content">It creates a parse tree from page source code that can be used to extract data easily.</p>
</body>
</html>
"""

# HTML ÌååÏã±
soup = BeautifulSoup(html_doc, 'html.parser')

# ÌäπÏ†ï ÌÉúÍ∑∏ ÎÇ¥Ïö© Í∞ÄÏ†∏Ïò§Í∏∞
title = soup.title
h1 = soup.h1
paragraphs = soup.find_all('p')

# ÎÇ¥Ïö© Ï∂úÎ†•
print("Title:", title.text if title else "No title found")
print("H1:", h1.text if h1 else "No h1 found")
for p in paragraphs:
    print("Paragraph:", p.text)
```

```python
import requests
from bs4 import BeautifulSoup

url = "https://search.daum.net/search?w=tot&q=2022%EB%85%84%EC%98%81%ED%99%94%EC%88%9C%EC%9C%84&DA=MOR&rtmaxcoll=MOR"

response = requests.get(url)
response.raise_for_status()  # Ïò§Î•ò Î∞úÏÉù Ïãú ÏòàÏô∏ Ï≤òÎ¶¨

html_content = response.text
soup = BeautifulSoup(html_content, "html.parser")

rankings = soup.find_all(class_ = "type_plural list_exact movie_list") #"type_plural list_exact movie_list" ÌÅ¥ÎûòÏä§ Î™®Îëê rankingsÏóê Îã¥Í∏∞



for ranking in rankings: # Ï≤´Î≤àÏß∏ forÎ¨∏ÏùÄ "type_plural list_exact movie_list"Í∞Ä Î™®Îëê ÏûàÎäî rankingsÏóêÏÑú Ï†úÎ™©, ÎàÑÏ†Å, ÌèâÏ†êÏùò ÌÅ¥ÎûòÏä§Îßå ÎΩëÍ∏∞ ÏúÑÌï¥ ÏÇ¨Ïö©
    movie_titles = ranking.find_all(class_= "info_tit")
    accumulate = ranking.select(".dl_comm:nth-of-type(4)")
    grade = ranking.find_all(class_ = "dl_comm star_comp_s")
    year_acc = ranking.select(".dl_comm:nth-of-type(3)")
    release = ranking.select(".dl_comm:nth-of-type(2)")

    
    

    for index, (title,acc,gr,year,rel) in enumerate(zip(movie_titles,accumulate,grade,year_acc,release,), start=1):    # ÎëêÎ≤àÏß∏ forÎ¨∏ÏùÄ Ï†úÎ™©, ÎàÑÏ†Å, ÌèâÏ†êÏùò ÌÅ¥ÎûòÏä§ ÏïàÏóêÏÑú ÏàúÏúÑÎ•º Ï†ïÌïòÍ∏∞ ÏúÑÌïú forÎ¨∏

        print(f"{index}){title.text}\n{gr.text}\n{year.text}\n{acc.text}\n{rel.text}")
      
    

```

```python
#Ïù¥ÎØ∏ÏßÄ Ï†ÄÏû• 1ÎÖÑ

import requests
from bs4 import BeautifulSoup
import urllib

url = "https://search.daum.net/search?w=tot&q=2022%EB%85%84%EC%98%81%ED%99%94%EC%88%9C%EC%9C%84&DA=MOR&rtmaxcoll=MOR"

response = requests.get(url) # url ÏöîÏ≤≠
response.raise_for_status()  # ÏöîÏ≤≠Ïù¥ Ïã§Ìå®Ìïú Í≤ΩÏö∞ ÏóêÎü¨ Ï≤òÎ¶¨

html_content = response.text # responseÏùò ÌÖçÏä§Ìä∏Î•º Î™®Îëê Î∂àÎü¨Ïò¥
soup = BeautifulSoup(html_content, "html.parser")

# Ïù¥ÎØ∏ÏßÄ ÌÉúÍ∑∏ Î™®Îëê Ï∞æÍ∏∞
image_tags = soup.find_all(class_="thumb_img")

# ÌååÏùºÏù¥ Ï†ÄÏû•Îê† Í≤ΩÎ°ú ÏßÄÏ†ï
directory = '/Users/choejong-gyu/Desktop/image/'  # Ïù¥ÎØ∏ÏßÄÍ∞Ä Ï†ÄÏû•Îê† ÎîîÎ†âÌÜ†Î¶¨Î™Ö

file_no = 0
for img_tag in image_tags:
    img_url = img_tag['src'] # srcÏùò ÏÜçÏÑ±Ïùò Í∞íÏùÑ Í∞ÄÏ†∏Ïò¥.
    try:
        filename = f"{directory}/image{file_no}.jpg"  # ÌååÏùºÎ™Ö ÏßÄÏ†ï
        urllib.request.urlretrieve(img_url, filename) # Ïù¥ÎØ∏ÏßÄ Ï†ÄÏû•ÌïòÎäî Ìï®Ïàò
        file_no += 1
    except Exception as e:
        print(f"Error occurred: {e}") #ÏóêÎü¨ÎÇ¨ÏùÑ Îïå
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
import requests
import pandas as pd

service = Service()
options = webdriver.ChromeOptions()
driver = webdriver.Chrome(service=service, options=options)
driver.get("https://www.coupang.com/np/categories/194276")
list_elements = driver.find_elements(By.CLASS_NAME, 'baby-product-link')

file_no = 0
directory = '/Users/choejong-gyu/Desktop/image/'  # ÏõêÌïòÎäî ÎîîÎ†âÌÜ†Î¶¨Î°ú Î≥ÄÍ≤ΩÌïòÏÑ∏Ïöî


data_list = []

# ÌÖçÏä§Ìä∏ Î∞è Ïù¥ÎØ∏ÏßÄ ÌÅ¨Î°§ÎßÅ
for index, element in enumerate(list_elements, start=1):
    product_name = element.find_element(By.CLASS_NAME, "name")
    price = element.find_element(By.CLASS_NAME, "price-value")
    rate = element.find_element(By.CLASS_NAME, "rating")
    count = element.find_element(By.CLASS_NAME, "rating-total-count")
    
    # Ïù¥ÎØ∏ÏßÄ Ï†ÄÏû•
    try:
        img_tag = element.find_element(By.TAG_NAME, "img")
        img_url = img_tag.get_attribute("src")
        print(f"{index}) {product_name.text}\n{price.text}Ïõê\n{rate.text}Ï†ê\nÌèâÏ†ê{count.text}Í∞ú")
        print(f"Ïù¥ÎØ∏ÏßÄ URL: {img_url}")

        if img_url and img_url.startswith("http"):  # URLÏù¥ ÎπÑÏñ¥ ÏûàÏßÄ ÏïäÍ≥† "http"Î°ú ÏãúÏûëÌïòÎäîÏßÄ ÌôïÏù∏
            filename = f"{directory}/{file_no+1}Îì±,{product_name.text}.jpg"
            with open(filename, "wb") as img_file:
                img_file.write(requests.get(img_url).content)
            file_no += 1
        else:
            print("ÏûòÎ™ªÎêú Ïù¥ÎØ∏ÏßÄ URL ÎòêÎäî Îπà URL")
        data_list.append({
            'Ïù¥Î¶Ñ': product_name.text,
            'Í∞ÄÍ≤©': price.text,
            'ÌèâÏ†ê': rate.text,
            'URL': img_url,
            'ÌååÏùºÏù¥Î¶Ñ': filename
        })
    except Exception as e:
        print(f"Ïò§Î•ò Î∞úÏÉù: {e}")
        

final = pd.DataFrame(data_list)
final.to_csv('Ïø†Ìå°ÏãùÌíà.csv', encoding='utf-8-sig')
final      
```
